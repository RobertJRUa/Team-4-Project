```python
# Import dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json 
import requests
import time
#from citipy import citipy

# Import api key 
from api_keys import gkey 
```

```python
# Build base url for text search 
text_type = "lodging"
location = "hotels in Cuyahoga County"

text_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?"

params = {"type" : text_type,
         "query" : location,
         "key" : gkey}

response = requests.get(text_url, params=params)
cuyahoga_text = response.json()

print(json.dumps(cuyahoga_text, indent=4, sort_keys=True))
```

```python
# Show response url 
response.url
```

```python
# Create lists to hold all of our data points
address = []
lat = []
lng = []
name = []
place_id = []
rating = []
user_ratings_total = []
types = []
```

```python
# Building lists 
for element in cuyahoga_text['results']:
    address.append(element['formatted_address'])
    lat.append(element['geometry']['location']['lat'])
    lng.append(element['geometry']['location']['lng'])
    name.append(element['name'])
    place_id.append(element['place_id'])
    rating.append(element['rating'])
    user_ratings_total.append(element['user_ratings_total'])
    types.append(element['types'])
```
